# Model Comparison

```python
time = np.arange(N_D.shape[1]) * 0.1
dNdt = hlp.cubic_splines(N_D[1],time); dNovN = dNdt / N_D[1]
true = dNovN
mean = np.mean(dNovN,axis=(1,2))
```

```python
model_list = [fs1MB,fs1E,fs2MB,fs2E,fs3MB,fs3E,fs4MB,fs4E,fs5MB,fs5E,
              fs9MB,fs9E,fs10MB,fs10E,fs11MB,fs11E,fs12MB,fs12E,fsTDMB,fsTDE,fs10HP]

model_errors = {"MB1":[0],"BS1":[1],"MB2":[2],"BS2":[3],"MB3":[4],"BS3":[5],
                "MB4":[6],"BS4":[7],"MB5":[8],"BS5":[9],"MB9":[10],"BS9":[11],
                "MB10":[12],"BS10":[13],"MB11":[14],"BS11":[15],"MB12":[16],"BS12":[17],
                "MBTD":[18],"BSTD":[19],"HP10":[20]}

for model in model_errors:
    preds = hlp.predict(model_list[model_errors[model][0]]['equation'],DIFF_DF[0][1])
    model_errors[model].append(hlp.get_errors(true,preds,agg="Space"))

null_error = hlp.Null_Model_Errors(true,mean).mean(axis=0)
```

```python
BSEQs = pd.concat([fs1E,fs2E,fs3E,fs4E,fs5E,fs9E,fs10E,fs11E,fs12E,fsTDE],axis=1).T; BSEQs.reset_index(drop=True, inplace= True)
MBEQs = pd.concat([fs1MB,fs2MB,fs3MB,fs4MB,fs5MB,fs9MB,fs10MB,fs11MB,fs12MB,fsTDMB],axis=1).T; 
MBEQs.reset_index(drop=True, inplace= True)
BSEQs["Type"] = "Best Score"
MBEQs["Type"] = "Model Best"
HPEQ = fs10HP; HPEQ["Type"] = "Hand Pick"
BEQs = pd.concat([BSEQs,MBEQs],axis=0); BEQs.reset_index(drop=True, inplace=True);
BEQs = pd.concat([BEQs,pd.DataFrame(HPEQ).T],axis=0); BEQs.reset_index(drop=True, inplace=True)
```

```python
fig, ax = plt.subplots()
inds = [5,15,18,16,19,20]
BEQSP = BEQs.iloc[inds]
noise = np.random.normal(0,0.05,len(BEQSP)) 
sns.scatterplot(data=BEQSP, x=BEQSP['complexity']+noise, y='loss', hue='features', style='Type',s=150, ax=ax)
sns.move_legend(ax, bbox_to_anchor=(1, 0.5), loc='center left')
ax.set_axisbelow(True)
ax.grid(color='gray', linestyle='dashed')
ax.set_title("Diffusion Model Comparison")
ax.set_ylabel("training loss")
plt.show()
```

# Interpreting the Equation

```python
HPeq10 = '(np.exp((-9.511543e-8 * Cum_N) + -0.7169847) + np.exp((Cum_Nbar / -0.12202544) / Cum_N))'
```

```python
MBeq10 = '(np.exp((-9.511543e-8 * Cum_N) + -0.7140626) + ((-0.019496204 * (Cum_Nbar - Cum_N)) / (Cum_Nbar + (-0.078773394 / -9.783274e-8))))'
```

```python
mx1 = 0; mx2 = 0;
for row in DIFF_DF[0][1]:
    for df in row:
        val1 = df['Cum_N'].max()
        val2 = df['Cum_Nbar'].max()
        if val1 > mx1: mx1 = val1
        if val2 > mx2: mx2 = val2
print(mx1,mx2)
```

```python
mn1 = mx1; mn2 = mx2;
for row in DIFF_DF[0][1]:
    for df in row:
        val1 = df['Cum_N'].min()
        val2 = df['Cum_Nbar'].min()
        if val1 < mn1: mn1 = val1
        if val2 < mn2: mn2 = val2
print(mn1,mn2)
```

```python
x = np.linspace(mn1-10,mx1+5e7,1000)
y = np.linspace(mn1-10,mx1-5e7,1000)
Cum_N, Cum_Nbar = np.meshgrid(x,y)
preds = eval(HPeq10,None,{"Cum_N":Cum_N,"Cum_Nbar":Cum_Nbar})
preds2 = eval(MBeq10,None,{"Cum_N":Cum_N,"Cum_Nbar":Cum_Nbar})
```

```python
import matplotlib.colors as colors

fig, axs = plt.subplots(figsize=(8,6))

hm = axs.pcolor(Cum_N,Cum_Nbar,preds,
           norm=colors.LogNorm(vmin=np.min(preds), vmax=np.max(preds)),cmap="YlGn"
          )

colony_traces = [(0,0),(9,9),(31,47),(1,1),(10,0)]
c = DIFF_DF[0][1][0][0]["time"].values

for (i,j) in colony_traces:
    trace = DIFF_DF[0][1][i][j]
    scat = axs.scatter(trace['Cum_N'],trace['Cum_Nbar'],c=c,cmap="RdBu",s=10)
    x,y = scat.get_offsets()[-1]
    if (i,j) == (0,0): axs.annotate(f"{(i,j)}", xy=(x,y+8e6),fontsize=8)
    else: axs.annotate(f"{(i,j)}", xy=(x+8e6,y+8e6),fontsize=8)



axs.set_xlabel(r"$N_C$"); axs.set_ylabel(r"$\bar{N}_C$")
axs.set_title(r"$\frac{dN}{dt\cdot N} = f(N_C,\bar{N}_C)$")

fig.tight_layout()
fig.colorbar(hm,ax=axs)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(8,6))

hm = axs.pcolor(Cum_N,Cum_Nbar,preds2,
          norm=colors.LogNorm(vmin=np.min(np.abs(preds2)), vmax=np.max(preds2)),cmap="YlGn")
                  
colony_traces = [(0,0),(9,9),(31,47),(1,1),(10,0)]
c = DIFF_DF[0][1][0][0]["time"].values

for (i,j) in colony_traces:
    trace = DIFF_DF[0][1][i][j]
    scat = axs.scatter(trace['Cum_N'],trace['Cum_Nbar'],c=c,cmap="RdBu",s=10)
    x,y = scat.get_offsets()[-1]
    if (i,j) == (0,0): axs.annotate(f"{(i,j)}", xy=(x,y+8e6),fontsize=8)
    else: axs.annotate(f"{(i,j)}", xy=(x+8e6,y+8e6),fontsize=8)

axs.set_xlabel(r"$N_C$"); axs.set_ylabel(r"$\bar{N}_C$")
axs.set_title(r"$\frac{dN}{dt\cdot N} = f(N_C,\bar{N}_C)$")

fig.tight_layout()
fig.colorbar(hm,ax=axs)
plt.show()
```
